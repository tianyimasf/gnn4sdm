```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
# url = 'https://api.globalbioticinteractions.org/locations'
with open('../data/species_interaction_locations.json', 'r') as file:
    locations = json.load(file)["data"]

print("location successfully loaded.")
print("first location", locations[0])
```

```python
bounding_box1 = [39.074546,-96.666819]
bounding_box2 = [39.147588,-96.544617]

from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers

    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c  # Distance in kilometers
    
def find_nearest_location(lat, lon):
    distances = [haversine_distance(location[0], location[1], lat, lon) for location in locations]
    idx = np.argmin(distances)
    return locations[idx]

location1 = find_nearest_location(bounding_box1[0], bounding_box1[1])
location2 = find_nearest_location(bounding_box2[0], bounding_box2[1])

print(location1, location2)
```

```python
c = 10
url = f'https://api.globalbioticinteractions.org/interaction?bbox={location1[0]-c},{location1[1]+c},{location2[0]-c},{location2[1]+c}'
print(url)
response = requests.get(url)
print(response.text)
```
